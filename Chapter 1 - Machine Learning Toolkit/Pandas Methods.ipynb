```python
import pandas as pd
import numpy as np

df = pd.read_csv('titanic.csv')
```

We have already seen the **head** method, which returns the first 5 lines of the DataFrame, we can also select more or less lines:

```python
df.head(n=7)
```

```python
df.head(n=2)
```

### Descriptive Statistics

We can also compute summary statistics for numerical columns in the *DataFrame*:

```python
df.describe()
```

We can also compute the mean, standard deviation, min, max separately

```python
df.mean()
```

```python
df.std()
```

```python
df.min()
```

```python
df.max()
```

```python
df.quantile(0.75)
```

But what about the columns of text information?  We haven't got any descriptive statistics.  By default **describe** only includes numerical data, what we can get some information on text columns by passing the argument **include='all'**:

```python
df.describe(include='all')
```

```python
type(df.describe(include='all'))
```
